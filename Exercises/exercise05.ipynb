```python
# 5 a 
import random
outcome_dice_rolls = [random.randint(1, 6) for i in range(100)]
count = 0
for outcome_dice_roll in outcome_dice_rolls:
    if outcome_dice_roll == 6:
        count += 1
print(f"The total number of outcome of six is {count}")
```

```python
# 5
dice_rolls = [10, 100, 1000, 10000, 100000, 1000000]

outcome_dice_rolls = []

probabilities = []
counts = []


for i in range(len(dice_rolls)):
    outcome = []
    # generate random numbers in range of number in position i
    for j in range(dice_rolls[i]):
        outcome.append(random.randint(1, 6))
    outcome_dice_rolls.append(outcome)
    count = 0
    # Count the number of six in each loop
    for h in range(len(outcome)):
        if outcome[h] == 6:
            count += 1
    counts.append(count)
    probabilities.append(round(count / dice_rolls[i], 2))

for probability in probabilities:
    position = probabilities.index(probability)
    print(f"The probability of getting six when you roll the dice for {dice_rolls[position]} times is: {probability * 100}")

plt.plot(probabilities)
plt.show()
```

```python
import math
points = [[random.uniform(-1, 1), random.uniform(-1, 1)] for i in range(5000)]
points_distances = []

for point in points:
    distance_from_origo = math.sqrt(point[0] ** 2 + point[1] ** 2)
    points_distances.append(distance_from_origo)

orange_team = []
blue_team = [] 

for point_distance in points_distances:
    if point_distance > 1: 
        blue_team.append(point_distance)
    else:
        orange_team.append(point_distance)
    
plt.plot(orange_team, '.')
plt.plot(blue_team, '.')

plt.show()
```
